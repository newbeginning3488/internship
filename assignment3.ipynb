```python
urls = []    
for i in range(0,2):  
    page_url = driver1.find_elements_by_xpath("//a[@class='a-link-normal s-no-outline']")
    for i in page_url:
        urls.append(i.get_attribute("href"))
        next_btn = driver1.find_element_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
        
```

```python
len(urls)
```

```python


Brand_Name = []
Product_Name = []
Rating = [] 
No_of_Ratings = []
Price = []
Exchange = []
Expected_delivery = []
Availability=[]
Other_Details = []

for i in urls:
    driver1.get(i)
    
    
    try:
        brand = driver1.find_element_by_xpath("//a[@id='bylineInfo']")
        Brand_Name.append(brand.text)
    except NoSuchElementException:
        Brand_Name.append('-')
    
    try:
        name = driver1.find_element_by_xpath("//span[@id='productTitle']")
        Product_Name.append(name.text)
    except NoSuchElementException:
        Product_Name.append('-')
    
    try:
        ratings = driver1.find_element_by_xpath("//span[@class='a-size-base a-nowrap']/span")
        Rating.append(ratings.text)
    except NoSuchElementException:
        Rating.append('-')
       
    try:
        no_of_ratings = driver1.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        No_of_Ratings.append(no_of_ratings.text)
    except NoSuchElementException:
        No_of_Ratings.append('-') 
    
    try:
        price = driver1.find_element_by_xpath("//td[@class='a-price']")
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')
    
    try:
        Return = driver1.find_element_by_xpath("//span[@class='a-declarative']/div/a")
        Exchange.append(Return.text)
    except NoSuchElementException:
        Exchange.append('-')
        
    try:
        exp_del = driver1.find_element_by_xpath("//div[@class='a-spacing-base']/b")
        Expected_delivery.append(exp_del.text)
    except NoSuchElementException:
        Expected_delivery.append('-')
    
    try:
        availability = driver1.find_element_by_xpath("//span[@class='a-size-base a-color-success a-text-bold']")
        Availability.append(availability.text)
    except NoSuchElementException:
        Availability.append('-')
        
    try:
        other_details = driver1.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']")
        Other_Details.append(other_details.text)
    except NoSuchElementException:
        Other_Details.append('-')
```

```python
len(Brand_Name),len(Product_Name),len(Rating),len(No_of_Ratings),len(Price),len(Exchange),len(Expected_delivery),len(Availability),len(Other_Details)
```

```python
Product = pd.DataFrame({})
Product["Brand Name"] = Brand_Name
Product["Product Name"] = Product_Name
Product["Rating"] = Rating
Product["No Of Ratings"] = No_of_Ratings
Product["Price"] = Price
Product["Return/Exchange"] = Exchange
Product["Expected Delivery"] = Expected_delivery
Product["Availability"] = Availability
Product["Other Details"] = Other_Details
Product["Product URL"] = urls
Product
```
