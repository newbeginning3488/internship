# write a python program to display IMDB'S toprated 100 movie's data (i.e. name,rating,year of release) and make a DataFrame

```python
# importing the required libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
url = "https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc"
```

```python
page = requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
```

```python
page
```

```python
soup = BeautifulSoup(page.content,"html.parser")
```

```python
movie_name = []
year = []
rating = []
```

```python
movie_data = soup.findAll('div',attrs={'class': "lister-item mode-advanced"})
```

```python
for store in movie_data:
    name = store.h3.a.text
    movie_name.append(name)
    
```

```python
movie_name
```

```python
for store in movie_data:
    year_of_release = store.h3.find('span',class_="lister-item-year text-muted unbold").text.replace('(','').replace(')','')
    year.append(year_of_release)
                                                                                                                     
```

```python
year
```

```python
for store in movie_data:
    rate = store.find('div',class_="inline-block ratings-imdb-rating").text.replace('\n','')
    rating.append(rate)
```

```python
rating
```

```python
np.count_nonzero(rating)
```

```python
movie_DF = pd.DataFrame({'Name of movie': movie_name, 'Year of release': year, 'Movie Rating': rating })
```

```python
movie_DF
```

```python
movie_DF.to_csv('IMDB Top Movies.csv', index=False)
```
