```python
driver5= webdriver.Chrome(r"C:\Users\reach\chromedriver.exe")
```

```python
url = 'https://github.com/' 
driver5.get(url)
```

```python
# Getting explore button and clicking on it
explore = driver5.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details").click()

#Selecting trending option
trend_url = driver5.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a")
urls = trend_url.get_attribute("href")
driver5.get(urls)
```

```python
URLs = []
repository_title = []
Description = []
Contributors = []
Language = []
lang = []
```

```python
#Fetching urls for each repository
repository = driver5.find_elements_by_xpath("//h1[@class = 'h3 lh-condensed']//a")
for i in repository:
    URLs.append(i.get_attribute("href"))

    
# Scraping Repository Title data
title = driver5.find_elements_by_xpath("//h1[@class = 'h3 lh-condensed']")
for i in title:
    repository_title.append(i.text)

    
# Scraping data from all repository page
for i in URLs:
    driver5.get(i)
    
    
    
    # Scraping Repository Description data
    try:
        desc = driver5.find_element_by_xpath("//p[@class='f4 mt-3']")
        Description.append(desc.text)
    except NoSuchElementException:
        Description.append('-')
        
        # Scraping Contributors Count data
    try:
        contributor = driver5.find_element_by_xpath("//*[contains(text(),'    Contributors ')]")
        Contributors.append(contributor.text.replace('Contributors',''))
    except NoSuchElementException:
        Contributors.append('-')
            
    # Scraping Languages used data
    try:
        for i in driver5.find_elements_by_xpath("//span[@class='color-fg-default text-bold mr-1']"):
            lang.append(i.text)
        Language.append(lang)
    except NoSuchElementException:
        Language.append('-')
```

```python
Github=pd.DataFrame({})
Github['Repository title'] = repository_title
Github['Repository description'] = Description
Github['Contributors count'] = Contributors
Github['Language used'] = Language
Github
```
