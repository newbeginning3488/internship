# womens cricket along with matches,team and rating

```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
url = 'https://www.icc-cricket.com/rankings/womens/team-rankings/odi'
```

```python
r = requests.get(url)
```

```python
soup = bs(r.text,'html.parser')
```

```python
table  = soup.find('table', class_="table")
body = table.find('tbody')

tr = body.find_all('tr')
row = []

# looping through the tr tag to extract the data
for i in tr:
  row_data = i.find_all('td')
  l = []
  for j in row_data:
    p = j.text.strip().split('\n')
    
    l.append(p)

  row.append(l)
```

extracting data from the row

```python
list = []
for i in row:
  #print(i)
  pos = i[0][0]
  team = i[1][0]
  matches = i[2][0]
  point = i[3][0]
  rating = i[4][0]
  
  #---------------------------------------Saving the data into list of dictionary--------------------------------
  d = {
      'POS' : pos,
       'TEAM' : team,
       'MATCHES' : matches,
       'POINTS' : point,
       'RATING' : rating
       }
  list.append(d)
```

```python
df = pd.DataFrame(list)
df
```
